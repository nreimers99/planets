```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/home/2022_07_13_WTvsKO_allcompounds_neg.csv")
```

```python
data
```

```python
retention_time = data['Unnamed: 4'][2:].to_numpy().astype('float32')
ccs = data['Unnamed: 5'][2:].to_numpy().astype('float32')
mz = data['Unnamed: 2'][2:].to_numpy().astype('float32')
mz = mz[:1000]
retention_time = retention_time[:1000]
ccs = ccs[:1000]
```

```python
# m/z conditions
truth_ls = []
for i in range(len(mz)):
  if i%100 == 0:
    print(i)
  mz_new = np.delete(mz,i)
  for j in range(len(mz_new)):
    if (mz_new[j] <= mz[i]+6.05 and mz_new[j] >= mz[i]+6.03) or (mz_new[j] <= mz[i]-6.05 and mz_new[j] >= mz[i]-6.03):
      truth_ls.append([i,j])
```

```python
len(truth_ls)
```

```python
# Retention time conditions
truth_ls_2 = []
for i in range(len(truth_ls)):
  idx1 = truth_ls[i][0]; idx2 = truth_ls[i][1]
  rt1 = retention_time[idx1]
  retention_time_new = np.delete(retention_time,idx1)
  rt2 = retention_time_new[idx2]
  if rt1 <= rt2+rt2*.03 and rt1 >= rt2-rt2*.03:
    truth_ls_2.append(truth_ls[i])
```

```python
len(truth_ls_2)
```

```python
# CCS conditions
truth_ls_3 = []
for i in range(len(truth_ls_2)):
  idx1 = truth_ls_2[i][0]; idx2 = truth_ls_2[i][1]
  rt1 = ccs[idx1]
  ccs_new = np.delete(ccs,idx1)
  rt2 = ccs_new[idx2]
  if rt1 <= rt2+rt2*.05 and rt1 >= rt2-rt2*.05:
    truth_ls_3.append(truth_ls_2[i])
```
